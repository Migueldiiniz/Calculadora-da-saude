```python
# Calculadora da saúde
```

```python
Peso = int(input('inserir peso: '))
altura = float(input('inserir altura: '))
```

```python
Escolha = int(input('Escolha uma opção:\n1. Calcular IMC\n2. Calcular Água\n3. Calcular peso ideal\n4. Gasto calórico diário\nSua escolha: '))
```

```python
if Escolha == 1:
   resultado = Peso / (altura * altura)
   print(f"O resultado é: {resultado:.2f}")
elif Escolha == 2:
   resultado = Peso * 35
   print(f"Sua meta é: {resultado / 1000:.2f} Litros de água")
elif Escolha == 3:
   resultado = (altura * 100) - 100
   print(f"Seu peso ideal estimado é: {resultado:.2f} kg")
elif Escolha == 4:
   resultado = (Peso * 10) + 1000
   print(f"Gasto calórico estimado (com treino): {resultado:.2f} kcal")
else:
   print('Não válido')
```
